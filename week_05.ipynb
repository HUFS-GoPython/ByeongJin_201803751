### 반복하기 : while
    
    - while 문은 가장 간단한 루핑(looping) 매커니즘이다.
    
    - while 조건문:
        실행코드 ...
        ...
        
        - 해당 조건이 거짓이면 코드를 실행하지 않는다. (언제 조건이 거짓이 되는지를 잘 파악해야 한다.)
        
    - 중단하기 : break
    
    - 건너뛰기 : continue
        - 파이썬에서 continue문은 반복문이 실행하는 코드블록의 나머지 부분을 실행하지 않고 다음 반복으로 건너가도록 흐름을 조정한다.
    
    - break 확인하기 : else
        - break 문은 어떤 내용을 확인하여 그것을 발견하면 종료하는 while 문을 작성할 때 사용한다.
        - while 문이 모두 실행되었지만 발견하지 못했을 때는 else 문이 실행된다.

```python
count = 1

while count <= 10: # count 가 10보다 작거나 같은 동안에 .. (조건)
    print(count) # 명령문을 실행한다.
    count += 1 # ..
```

```python
# break
count = 1
pw = int(input("번호를 입력하십시오."))

while True:
    
    if count == pw:
        
        print(F"설정한 숫자는 .. {count}")
        break # if 문의 조건이 성립하고 break 까지 도달했다면 더 이상 반복문을 돌아다니지 않는다.
    
    print(F"{count}... CHECK")
    count += 1
```

```python
# break .. else
# 조건문이 모두 실행되었지만 조건을 만족하지 못했을 때

count = 1
pw = int(input("숫자를 입력해 주십시오. 1 ~ 30"))

while count <= 30:
    
    if count == pw:
        
        print(F"PW는.. {count}")
        break
        
    print(F"{count} ... CHECK")
    count += 1
    
else: # 모든 조건 내에도 
    print("범위 내에 PW가 존재하지 않습니다.")
```

```python
# continue
count = 1

while count <= 10:
    
    if count % 2 != 0:
        
        count += 1
        continue # if 문의 조건이 성립하고 continue 까지 도달한 상태라면 나머지 코드를 실행하지 않고 시작으로 돌아간다.
    
    print(count)
    count += 1
```

### 순회하기 : for와 in

    - for 문에서 사용하는 순서열은 리스트, 튜플, 심지어 문자열 등 어떤 것을 사용해도 괜찮다. for 문은 순서열의 각 원소를 처음부터 차례로 순회하면서 반복 변수에 담아낸다.

    - 이터레이터(iterator : 반복자)는 유용하게 사용될 수 있다.
        - 자료구조가 얼마나 큰지, 어떻게 구현되었는지에 관계없이 자료구조를 순회할 수 있고, 심지어 바로 생성되는 데이터도 순회할 수 있다. 데이터가 메모리에 맞지 않더라도 데이터 스트림을 처리할 수 있도록 허용한다.
        - 리스트와 튜플, 딕셔너리와 같은 다른 순회 가능한(iterable) 항목에도 사용할 수 있다.
        
    - 중단하기 : break
        - for 문의 break 문은 while 문의 break 문과 똑같이 동작한다.
        
    - 건너뛰기 : continue
        - for 문의 continue 문은 while 문의 continue 문과 똑같이 동작한다.
        
    - break 확인하기 : else
        - for 문은 while 문과 같이 모든 항목을 순회했는지 확인하는 부가적인 옵션의 else 문. for 문에서 break 문이 호출되지 않으면 else문이 실행된다.
    
    - 숫자 시퀀스 생성하기 : range()
        - range() 함수는 리스트나 튜플 같은 자료구조를 생성하여 저장하지 않더라도 특정 범위 내에서 숫자 스트림을 반환한다.
        - 컴퓨터의 메모리를 전부 사용하지 않고, 프로그램의 충돌없이 아주 큰 범위를 생성할 수 있게 해준다.

```python
for i in ['a', 'b', 'c', 'd']:
# 변수 i 에는 반복될 때마다 리스트 ['a', 'b', 'c', 'd']의 '요소' 가 차례대로 '복사'된다.
# 리스트의 길이가 4 이므로, 이 for 문은 4번 반복을 수행한다.
    print(i)
```

```python
# for 문에 가장 많이 사용되는 순서열은 Range 이다.
# range(시작값, 멈춤값, 연속하는 두 수의 차) 를 매개 변수로써 써줄 수 있다.

for i in range(0, 10, 2):
    print(i)
    
print("")

for i in range(0, 5): # 연속하는 두 수의 차는 생략할 수 있고 생략시 기본값은 1이다.
    print(i)

print('')

for i in range(7): # 하나의 매개변수만 입력하면 그 숫자는 멈춤값(반복횟수)으로 인식한다.
    print(i)
```

### for 와 while
    - for: 반복 횟수가 정해져있고, 변하지 않는다. ex) 전체 학생의 성적을 산출할 때
        
    - while: 횟수는 정해지지 않고, 중요하지 않다. 단, 어떠한 조건이 만족 될 때에 명령을 실행한다. ex) 가위바위보

```python
word1 = 'prophet'
offset = 0

while offset < len(word1): # while 문은 '조건식이 성립할 동안' 코드를 실행한다.
    
    print(word1[offset])
    offset += 1

# while 문과 for in 문으로 같은 결과를 내지만 다른 코드.

print('')

for letter in word1: # 반면 for in 문은 자료구조의 '구성요소' 각각을 순회하는 것이다.
    # word1 이라는 자료 안의 구성요소 각각을 letter로 인식하고, 하나하나 순회한다.
    print(letter)
```

### 구구단
    - 입력을 받아서 사용한다.

    - 1. n단 출력
    - 2. 2-n단까지 출력
    - 3. n, m 모두 입력 받아서 출력
    - 4. 전체를 프로그램으로 만든다.
        - 메뉴: 1) n, 2) 2-n, 3) n-m

```python
#1

input_dan = int(input("단을 입력해 주십시오. (12단까지)\n"))

if input_dan > 12:
    
    print("최대 단은 12단입니다. 12단으로 설정합니다..")
    input_dan = 12

print(f"===== {input_dan}단 =====")

for i in range(9):
    print(f"{input_dan} X {i + 1} = ", input_dan * (i + 1))

print("================")
```

```python
#2

input_dan = int(input("구구단 표를 출력합니다. (2단 부터 12단까지)\n"))
if input_dan > 12:
    input_dan = 12
    print("12 초과의 숫자는 입력하실 수 없습니다.")

elif input_dan < 2:
    input_dan = 2
    print("2 미만의 숫자는 입력하실 수 없습니다.")

print(f"===== {input_dan}단 까지의 표 =====\n")

for j in range(input_dan - 1):
    
    print(f"===== {j + 2}단 =====")
    
    for i in range(9):
        
        print(f"{j + 2} X {i + 1} = ", (j + 2) * (i + 1))
        
    print('===============')
```

```python
#3

print("구구단 표를 출력합니다. (12단까지)\n")
start_dan = int(input("시작 단을 입력해주십시오."))
end_dan = int(input("끝 단을 입력해주십시오."))

if start_dan < 2:
    start_dan = 2
    print("2단 보다 작을 수 없습니다. 시작 단을 2단으로 설정합니다..")
    
elif start_dan > 12:
    start_dan = 12
    print("12단 보다 클 수 없습니다. 시작 단을 12단으로 설정합니다..")
    
    
if end_dan > 12:
    end_dan = 12
    print("12단 보다 클 수 없습니다. 끝 단을 12단으로 설정합니다..")
    
elif end_dan < 2:
    end_dan = 2
    print("2단 보다 작을 수 없습니다. 끝 단을 2단으로 설정합니다..")

if end_dan > start_dan:
    
    for i in range(end_dan - start_dan + 1):
        print(f'===== {start_dan + i}단 =====')
        
        for j in range(9):
            print(f"{start_dan + i} X {j + 1} =", (start_dan + i) * (j + 1))
        
        print('===============')
            
elif end_dan == start_dan:
    
    print(f'===== {start_dan}단 =====')
    
    for j in range(9):
        print(f"{start_dan} X {j + 1} =", (start_dan) * (j + 1))
        
    print('===============')
    
elif end_dan < start_dan: # 끝 단이 시작 단 보다 작아도 작은 수 -> 큰 수 순으로 실행
    
    for i in range(start_dan - end_dan + 1):
        print(f'===== {end_dan + i}단 =====')
        
        for j in range(9):
            print(f"{end_dan + i} X {j + 1} =", (end_dan + i) * (j + 1))
            
        print('===============')
```
